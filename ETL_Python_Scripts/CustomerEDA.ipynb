# Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univariate analysis and visualization 

plt.figure(figsize=(10,24))
for i,col in enumerate(numerical_cols): # i is storing Estimated Income at index 0 and Superannuation Savings at index 1 
    plt.subplot(8,3,i+1) # It have 3 arguement arg1= no of row arg2= no of column and arg3=index where to put that column in 1 row and 2 column plot, i+1 as i=0 but here in plot index of subplot mentioned as 1,2,3,4.
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.xticks(rotation=30, fontsize=8,ha='right') #ha basically saying tilt text in right side
plt.tight_layout() # when subplot is used, it fix x labels and title nicely so that overlapping doesnt happen
plt.show()

```

## Heatmaps

```python
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f") # We used annot=true because its corr value between two columns
plt.title("Correlation Matrix")
plt.show()
```

## Insights:

### =>Strong Positive Correlations:

1. Bank Deposits & Saving Accounts (0.75): Indicates customers with higher deposits also tend to maintain higher savings accounts.

2. Bank Deposits & Checking Accounts (0.84): Customers with large deposits are also active in checking accounts.

3. Checking Accounts & Saving Accounts (0.75): Suggests customers commonly use both types of accounts together.

4. Bank Loans & Business Lending (0.42): Businesses taking loans may also be utilizing other lending facilities.

### => Weak to moderate correlation exists with estimated Income:

1. Superannuation Savings (0.37)

2. Credit Card Balance (0.30)

3. Bank Loans (0.33)

~Indicates income slightly influences financial activity, but not strongly.

### =>These show how different banking metrics are distributed among customers:
### >Right-Skewed Distributions (Majority with Lower Values):

1. Credit Card Balance, Bank Loans, Deposits, Business Lending, Foreign Currency, etc.

2. Indicates that a large portion of customers holds smaller amounts, and a few hold significantly high amounts.

3. Useful for targeting: High-value customers are rare but crucial for revenue.

### >Estimated Income:

1. Most customers earn between ₹100,000 and ₹250,000, with fewer earning above ₹400,000.

2. Suggests a mid-income customer base.

### >Superannuation Savings:

1. Concentrated below ₹40,000, meaning retirement planning is modest for most.

### =>Shows employee/customer count per job title:
### >Top Occupations:

1. Structural Analysis Engineer, Associate Engineer, HR Executive have the highest counts.

2. These professions may have stable income profiles, making them suitable banking customers for savings/loans.

### =>Business Insights for Banking Sector:

1. Focus on Cross-Selling:
Customers with checking accounts also likely have savings accounts → Bundle offers can work.

2. High-Value Targeting:
A small fraction holds large balances → Personalized services (e.g., wealth management) should be aimed at these segments.

3. Income Is Not the Sole Driver:
Moderate income correlation suggests behavioral or demographic segmentation could be more effective for marketing.

4. Occupation-Based Offers:
Tailor financial products (loans, savings) for top occupations (e.g., engineers and HR executives).

5. Loan-Linked Lending:
Those using bank loans are more inclined to use business lending—target SMEs or entrepreneurs for bundled credit services.
